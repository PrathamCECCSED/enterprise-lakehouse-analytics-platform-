```python
spark.sql("""
SELECT 
    video_category,
    COUNT(*) AS total_views,
    AVG(watch_time) AS avg_watch_time,
    SUM(engagement_score) AS total_engagement
FROM silver_youtube_events
GROUP BY video_category
ORDER BY total_engagement DESC
""").show()
```

```python
spark.sql("""
SELECT
    country,
    COUNT(*) AS total_views,
    SUM(engagement_score) AS engagement
FROM silver_youtube_events
GROUP BY country
ORDER BY engagement DESC
""").show()
```

```python
spark.sql("""
SELECT
    device_type,
    COUNT(*) AS total_views,
    AVG(watch_time) AS avg_watch_time
FROM silver_youtube_events
GROUP BY device_type
""").show()
```

```python
spark.sql("""
SELECT
    user_id,
    SUM(engagement_score) AS engagement
FROM silver_youtube_events
GROUP BY user_id
ORDER BY engagement DESC
LIMIT 10
""").show()
```
