```python
gold_df = spark.table("gold_category_kpis")
```

```python
gold_df.write.format("delta") \
    .mode("overwrite") \
    .partitionBy("video_category") \
    .saveAsTable("gold_category_kpis_partitioned")
```

```python
spark.sql("""
SELECT *
FROM gold_category_kpis_partitioned
LIMIT 5
""").show(truncate=False)
```

```python
spark.sql("""
DESCRIBE DETAIL gold_category_kpis_partitioned
""").show(truncate=False)
```
