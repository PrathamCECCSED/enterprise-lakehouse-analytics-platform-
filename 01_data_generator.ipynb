```python
import json
import random
from datetime import datetime

countries = ["India", "USA", "Canada", "UK", "Germany"]

devices = ["Mobile", "Desktop", "Tablet"]

categories = ["Gaming", "AI", "Tech", "Education", "Vlogs"]


def generate_event():

    event = {
        "event_time": str(datetime.now()),
        "user_id": random.randint(1000, 9999),
        "video_id": f"vid_{random.randint(1,50)}",
        "video_category": random.choice(categories),
        "country": random.choice(countries),
        "device_type": random.choice(devices),
        "watch_time": random.randint(30, 1200),
        "liked": random.choice([True, False]),
        "video_language": random.choice(
        ["English", "Hindi", "Spanish"]
    ),
        "subscribed": random.choice([True, False])
        
    }

    return event


events = []

for i in range(1000):
    events.append(generate_event())
    print(events[0])


file_name = f"/tmp/youtube_events_{datetime.now().date()}.json"

with open(file_name, "w") as file:
    json.dump(events, file)

print("File Generated Successfully")
print(file_name)
```

```python
import json
import random
import boto3
from datetime import datetime

countries = ["India", "USA", "Canada", "UK", "Germany"]

devices = ["Mobile", "Desktop", "Tablet"]

categories = ["Gaming", "AI", "Tech", "Education", "Vlogs"]


def generate_event():

    event = {
    "event_time": str(datetime.now()),
    "user_id": random.randint(1000, 9999),
    "video_id": f"vid_{random.randint(1,50)}",
    "video_category": random.choice(categories),

    "video_language": random.choice(
        ["English", "Hindi", "Spanish"]
    ),

    "country": random.choice(countries),
    "device_type": random.choice(devices),
    "watch_time": random.randint(30, 1200),
    "liked": random.choice([True, False]),
    "subscribed": random.choice([True, False])
}

    return event


events = []

for i in range(1000):
    events.append(generate_event())


file_name = f"/tmp/youtube_events_{datetime.now().date()}.json"

with open(file_name, "w") as file:
    json.dump(events, file)


AWS_ACCESS_KEY = ""
AWS_SECRET_KEY = ""

AWS_BUCKET_NAME = ""


s3 = boto3.client(
    "s3",
    aws_access_key_id="",
    aws_secret_access_key=""
)


s3.upload_file(
    file_name,
    ""
,
    f"raw/youtube_events_{datetime.now().date()}.json"
)

print("Daily YouTube events uploaded successfully")
```
